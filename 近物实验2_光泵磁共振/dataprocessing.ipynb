```python
import os

current_directory = os.getcwd()
all_contents = os.listdir(current_directory)

# 输出所有目录
print("所有目录：")
for content in all_contents:
    content_path = os.path.join(current_directory, content)
    if os.path.isdir(content_path):
        print(f"目录: {content}")

# 输出所有文件
print("\n所有文件：")
for content in all_contents:
    content_path = os.path.join(current_directory, content)
    if os.path.isfile(content_path):
        print(f"文件: {content}")
```

```python
import os

def list_bmp_files(directory):
    print(f"指定目录：{directory}")
    bmp_files = [file for file in os.listdir(directory) if file.lower().endswith('.bmp')]

    if not bmp_files:
        print("在指定目录下找不到任何 BMP 文件。")
    else:
        print("所有 BMP 文件：")
        for bmp_file in bmp_files:
            print(bmp_file)
    return bmp_files

def create_folder(directory, folder_name):
    new_folder_path = os.path.join(directory, folder_name)

    # 创建文件夹，包括不存在的中间目录
    os.makedirs(new_folder_path, exist_ok=True)

    print(f"已成功创建文件夹：{new_folder_path}")
    return new_folder_path

# # 指定位置的路径
# specified_directory = '/path/to/your/directory'

# # 新文件夹的名称
# new_folder_name = 'new_folder'

# # 调用函数创建新文件夹
# create_folder(specified_directory, new_folder_name)
```

```python
from PIL import Image

directory =dirlist[0]

# 调用函数列出所有 BMP 文件
bmplist=list_bmp_files(directory)

new_folder_path=create_folder(directory, 'inverted')
# 打开原始图片
for index,image_file in enumerate(bmplist):
    image = Image.open(os.path.join(directory, image_file))  # 你的原始图片路径
    pixels = list(image.getdata())
    inverted_pixels = [(255 - r, 255 - g, 255 - b) for (r, g, b) in pixels]
    width, height = image.size
    inverted_image = Image.new("RGB", (width, height))
    inverted_image.putdata(inverted_pixels)

    # 保存截取的图片
    # cropped_image.save("output_tiles/output_cropped_image.jpg")  # 保存路径可以根据需要修改

    inverted_image.save(f"{new_folder_path}/{index+1}_inverted_{image_file}.jpg")
    image.close()
    inverted_image.close()


```

```python
from PIL import Image

namelist=['FB','SJB','SC','CZC']
for adirlist,name in zip(dirlist,namelist):

    directory =adirlist

    # 调用函数列出所有 BMP 文件
    bmplist=list_bmp_files(directory)

    new_folder_path=create_folder(directory, 'inverted')
    # 打开原始图片
    for index,image_file in enumerate(bmplist):
        image = Image.open(os.path.join(directory, image_file))  # 你的原始图片路径
        pixels = list(image.getdata())
        inverted_pixels = [(255 - r, 255 - g, 255 - b) for (r, g, b) in pixels]
        width, height = image.size
        inverted_image = Image.new("RGB", (width, height))
        inverted_image.putdata(inverted_pixels)

        # 保存截取的图片
        # cropped_image.save("output_tiles/output_cropped_image.jpg")  # 保存路径可以根据需要修改

        inverted_image.save(f"{new_folder_path}/{name}_{index+1}.jpg")
        
        image.close()
        inverted_image.close()
    print('图片文件已经反色保存')
    print()

```
