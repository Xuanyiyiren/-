```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os

# 获取当前目录下的所有文件
file_list = [f for f in os.listdir() if os.path.isfile(f)]

# 按文件的最后修改时间进行排序，从上到下依次是从早到晚
sorted_file_list = sorted(file_list, key=lambda x: os.path.getmtime(x))

# 打印所有文件名
for file_name in sorted_file_list:
    if file_name[-3:]=='csv':
        print(file_name)
```

```python
# 读取Excel文件
df = pd.read_csv('zbjt1.csv')
```

```python
display(df)
```

```python
# 提取X轴和Y轴数据
x_data = df.iloc[1:, 0].to_numpy(dtype=float)
y_data = df.iloc[1:, 1].to_numpy(dtype=float)

dt=df['Increment'][0]
x_data*=dt

# x_data,y_data
# 创建折线图
# plt.scatter(x_data, y_data, linestyle='-')

plt.plot(x_data, y_data, marker='.', linestyle='-')
plt.xlabel('X轴标签')
plt.ylabel('Y轴标签')
plt.title('折线图标题')
plt.grid(True)

# 显示图表
plt.show()


```
