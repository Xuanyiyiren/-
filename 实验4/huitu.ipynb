```python
import sympy as sp

ome,tau=sp.symbols('omega,tau',real=True)

H=ome*tau/(1+(ome*tau)**2)**(5/2)
H
```

```python
expr=H.diff(ome)
expr
sp.solve(expr,ome)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['Simhei']  #显示中文
plt.rcParams['axes.unicode_minus']=False    #显示负号  

# 定义函数
def f(x):
    return x/(x**2+1)**(5/2)

# 生成x值
x = np.linspace(0, 4, 400)

# 计算y值
y = f(x)

# 找到函数的极值点
extrema_x =0.5  # 示例函数的极值点
extrema_y = f(extrema_x)

# 绘制函数曲线并标记极值点
plt.figure(figsize=(8,4))
plt.plot(x, y, label='幅度响应函数曲线',color='blue')
plt.scatter(extrema_x, extrema_y, color='red', marker='o', label='最大值')
plt.vlines(extrema_x, 0, extrema_y, linestyles='dashed', colors='red')
# 添加文字标注
plt.text(extrema_x+ 0.1, extrema_y, '极值,$\\omega=\\frac{1}{2\\tau}$', color='red')

# 添加图例和其他绘图元素
plt.legend()
plt.xlabel('$\\omega/{\\tau}^{-1}$')
plt.ylabel('$\\left| H(\\mathrm{i}\\omega)\\right|$')
plt.xlim(0,4)
plt.ylim(0,0.35)
plt.title('$RC-(CR)^4$ 滤波网络的幅度响应函数')
plt.grid(True)
plt.savefig('pic/fuduxiangying.pdf', bbox_inches='tight')
# 显示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义函数
def f(x):
    return x**3 - 4*x**2 + 4*x

# 生成x值
x = np.linspace(0, 4, 400)

# 计算y值
y = f(x)

# 找到函数的极值点
extrema_x = np.array([1.0, 2.0])  # 示例函数的极值点
extrema_y = f(extrema_x)

# 绘制函数曲线并标记极值点
plt.plot(x, y, label='Function Curve')
for i in range(len(extrema_x)):
    # 添加文字标注
    plt.annotate(f'({extrema_x[i]}, {extrema_y[i]})', (extrema_x[i], extrema_y[i] + 0.5), fontsize=12, color='red')
    # 绘制虚线从极值点到x轴
    plt.plot([extrema_x[i], extrema_x[i]], [0, extrema_y[i]], color='red', linestyle='--')
    plt.plot([0, extrema_x[i]], [extrema_y[i], extrema_y[i]], color='red', linestyle='--')

# 添加图例和其他绘图元素
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function Curve with Extrema Marked')
plt.grid(True)

# 显示图形
plt.show()
```
